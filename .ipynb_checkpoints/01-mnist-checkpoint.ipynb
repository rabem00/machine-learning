Keras is now a superset so you can do something like: import tensorflow.keras as keras

```python
import tensorflow as tf
print(tf.__version__)

#print("Num GPUs Available: ", 
#len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
mnist = tf.keras.datasets.mnist  # 28x28 images of handwritten digest 0-9
(x_train, y_train),(x_test, y_test) = mnist.load_data()  # load the data
```

```python
print(x_train[0])
```

```python
import matplotlib.pyplot as plt
```

```python
# Just take a peek
plt.imshow(x_train[0],cmap=plt.cm.binary)
plt.show()
```

```python
print(y_train[0])
```

```python
# normalize data between 0 and 1 (instead of 0 and 255)
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
# Just take a peek again
print(x_train[0])

plt.imshow(x_train[0],cmap=plt.cm.binary)
plt.show()
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu)) # 128 neurons or units, relu will fire on activation
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax)) # output layer
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3)
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)  # evaluate the out of sample data with model
print(val_loss)  # model's loss (error)
print(val_acc)  # model's accuracy
```

```python
model.save('num_reader.model')
```

```python
new_model = tf.keras.models.load_model('num_reader.model')
```

```python
predictions = new_model.predict(x_test)
```
